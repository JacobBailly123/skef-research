## 1. Mathematical Formulation

For a scalar field $h(x, z, t)$, we compute structural functions of the scalar Laplace PDE in two spatial variables $[x, z]$ and time $t$:

$$\frac{\partial^2}{\partial x^2} h(x,z,t) + \frac{\partial^2}{\partial z^2} h(x,z,t) = 0 \tag{1}$$

using a Stationary Kinematic Euler (SKE) structure with complex-valued parameters:

$$h(x,z,t) = H e^{Kx \cdot X + Kz \cdot Z} \tag{2}$$

where $H$ is a real structural coefficient, $[Kx, Kz] = [Lx + IMx,\; Lz + IMz]$ are complex structural parameters, $[X, Z] = [x - Vxt + Sx,\; z - Vzt + Sz]$ are propagation variables, $[Vx, Vz]$ are velocities, $[Sx, Sz]$ are spatiotemporal shifts, and $[x_0, z_0, t_0]$ are reference coordinates and time.

## 2. Computational Initialization & DIS Method

### Forward and Backward Substitutions

Setting up the scalar Laplace PDE and substitutions for the SKE structure:

$$h(x,z,t) = H \cdot eh \tag{2}, \quad eh = \frac{h(x,z,t)}{H} \tag{3}$$

$$eh = e^{KxX + KzZ} \tag{4}, \quad e^{KxX + KzZ} = eh \tag{5}$$

Propagation variables and spatiotemporal shifts:
$$X = x - Vx\,t + Sx, \quad Z = z - Vz\,t + Sz \tag{6}$$
$$X_n = x - x_0 - Vx(t - t_0), \quad Z_n = z - z_0 - Vz(t - t_0) \tag{8}$$

### Reduction of the Laplace PDE to the Laplace AE by the DIS Method

Spatial derivatives of the SKE structure using the chain rule:

$$\frac{\partial}{\partial x} h(x,z,t) = Kx \cdot h(x,z,t) \tag{10}$$
$$\frac{\partial}{\partial z} h(x,z,t) = Kz \cdot h(x,z,t) \tag{11}$$

Second spatial derivatives:

$$\frac{\partial^2}{\partial x^2} h(x,z,t) = Kx^2 \cdot h(x,z,t) \tag{13}$$
$$\frac{\partial^2}{\partial z^2} h(x,z,t) = Kz^2 \cdot h(x,z,t) \tag{14}$$

Substituting into the Laplace PDE yields the scalar Laplace AE:

$$\left(Kx^2 + Kz^2\right) h(x,z,t) = 0 \tag{15}$$

For this to hold for all $Kx$, $Kz$, and $h$, the structural coefficient must vanish:

$$Kx^2 + Kz^2 = 0 \tag{16}$$

This is the **complex structural algebraic equation** — the PDE is now an algebraic equation.


```python
# Numerical demonstration of the DIS reduction
# For h = H * e^(Kx*X + Kz*Z): d2h/dx2 + d2h/dz2 = (Kx^2 + Kz^2) * h

def h_ske(x, z, H, Kx, Kz, Vx=0, Vz=0, t=0, Sx=0, Sz=0):
    """SKE structural function h(x,z,t) = H * exp(Kx*X + Kz*Z)"""
    X = x - Vx*t + Sx
    Z = z - Vz*t + Sz
    return H * np.exp(Kx*X + Kz*Z)

def laplacian_num(f, x, z, h=1e-6):
    """Numerical Laplacian using finite differences"""
    d2dx2 = (f(x+h, z) - 2*f(x, z) + f(x-h, z)) / h**2
    d2dz2 = (f(x, z+h) - 2*f(x, z) + f(x, z-h)) / h**2
    return d2dx2 + d2dz2

# Test with Kx^2 + Kz^2 = 0 (complex structural AE satisfied)
# e.g. Kx = 1, Kz = I (purely imaginary) -> Kx^2 + Kz^2 = 1 + (-1) = 0
import cmath
Kx_c, Kz_c = 1.0, 1j  # satisfies Kx^2 + Kz^2 = 0

print("DIS method: structural AE Kx^2 + Kz^2 = 0")
print(f"  Kx = {Kx_c}, Kz = {Kz_c}")
print(f"  Kx^2 + Kz^2 = {Kx_c**2 + Kz_c**2} ✓")
print()

# Verify real SKEF functions satisfy Laplace PDE numerically
# Using reduced form: f(x,z,t) = exp((-1)^j * Rzn * z) * cos(Rzn * X)
j = 1  # upper domain
Rzn = 2.0
t_val, Vxn, Sxn = 1.0, 1.0, 1.0

def f_skef(x, z, Rzn=2.0, j=1, Vxn=1.0, Sxn=1.0, t=1.0):
    Xn = x - Vxn*t + Sxn
    Zn = (-1)**j * Rzn * z
    return np.exp(Zn) * np.cos(Rzn * Xn)

test_points = [(0.5, 0.1), (1.0, 0.2), (0.3, 0.15), (-0.2, 0.05)]
print("Verifying SKEF functions satisfy Laplace PDE (∇²f = 0):")
print(f"{'Point (x,z)':<20} {'∇²f (≈ 0)'}")
print("-" * 40)
for pt in test_points:
    lap = laplacian_num(lambda x,z: f_skef(x, z), pt[0], pt[1])
    print(f"  ({pt[0]}, {pt[1]}){'':<12} {lap:.2e}")
print("\nResult: SKEF functions satisfy Laplace's equation numerically.")
```

## 3. Complex-Valued SKEF Parameters

Representing complex parameters $[Kx, Kz]$ through real parameters:

$$Kx = Lx + I \cdot Mx, \quad Kz = Lz + I \cdot Mz \tag{17}$$

Substituting into the complex structural AE $Kx^2 + Kz^2 = 0$ and separating real and imaginary parts yields two real structural AEs:

**Condition of length invariance:**
$$Lx^2 + Lz^2 - Mx^2 - Mz^2 = 0 \tag{18}$$

**Orthogonality condition:**
$$Lx \cdot Mx + Lz \cdot Mz = 0 \tag{19}$$

These are separated by introducing polar coordinate systems for $[Lx, Lz]$ and $[Mx, Mz]$:

$$Lx = Rz\cos(A), \quad Lz = Rz\sin(A) \tag{20}$$
$$Mx = Rz\cos(\alpha), \quad Mz = Rz\sin(\alpha) \tag{21}$$

The orthogonality condition $\cos(A - \alpha) = 0$ is satisfied when:

$$A = \alpha + \frac{1}{2}\pi + \pi j \tag{27}$$

Substituting back, the first radius-vector becomes:

$$Lx = -(-1)^j Rz\sin(\alpha), \quad Lz = (-1)^j Rz\cos(\alpha) \tag{28}$$


```python
# Verify the length invariance and orthogonality conditions using polar coordinates
Rz = 2.5
alpha = np.pi / 4
j_vals = [0, 1]

print("Structural algebraic equations — polar coordinate substitution:")
print()
for j in j_vals:
    Lx = -(-1)**j * Rz * np.sin(alpha)
    Lz = (-1)**j * Rz * np.cos(alpha)
    Mx = Rz * np.cos(alpha)
    Mz = Rz * np.sin(alpha)
    
    length_inv = Lx**2 + Lz**2 - Mx**2 - Mz**2
    ortho = Lx*Mx + Lz*Mz
    
    print(f"  j = {j}:")
    print(f"    [Lx, Lz] = [{Lx:.4f}, {Lz:.4f}]")
    print(f"    [Mx, Mz] = [{Mx:.4f}, {Mz:.4f}]")
    print(f"    Eq. 18 (length invariance): {length_inv:.2e}")
    print(f"    Eq. 19 (orthogonality):     {ortho:.2e}")
    print()
```

## 4. Real-Valued SKEF Functions and Structures

Using the Euler formula $e^{a+Ib} = e^a(\cos(b) + I\sin(b))$ and substituting the structural parameters, the complex solution decomposes into real structural functions:

$$h(x,z,t) = H(f(x,z,t) + I\,g(x,z,t)) \tag{30}$$

**Real structural functions (general case):**
$$f(x,z,t) = e^{(-1)^j Rz(-\sin(\alpha)X + \cos(\alpha)Z)} \cos\!\left(Rz(\cos(\alpha)X + \sin(\alpha)Z)\right) \tag{31}$$
$$g(x,z,t) = e^{(-1)^j Rz(-\sin(\alpha)X + \cos(\alpha)Z)} \sin\!\left(Rz(\cos(\alpha)X + \sin(\alpha)Z)\right)$$

**Reduced structural functions** (polar axis along x-axis, i.e. $\alpha=0$, propagation in x-direction):

$$f(x,z,t) = e^{(-1)^j Rz\,z} \cos(Rz\,X_n) \tag{33}$$
$$g(x,z,t) = e^{(-1)^j Rz\,z} \sin(Rz\,X_n)$$

The **fundamental SKEF structure** $ah_n$ is:

$$ah_n(x,z,t) = \left(Fh_n\,cx_n + Gh_n\,sx_n\right) ez_n \tag{48}$$

where: $cx_n = \cos(Rz_n X_n)$, $sx_n = \sin(Rz_n X_n)$, $ez_n = e^{(-1)^j Rz_n z}$

The local Cartesian coordinate and decay variable:
$$X_n = x - Vx_n\,t + Sx_n \tag{45}, \quad Z_n = (-1)^j Rz_n\,z \tag{46}$$


```python
# Fundamental SKEF structure ah_n (Eq. 48)

def ah_n(x, z, t, Rzn, Vxn, Sxn, Fhn, Ghn, j=1):
    """
    Fundamental SKEF structure (Eq. 48):
    ah_n = (Fhn*cx_n + Ghn*sx_n) * ez_n
    j=1: upper domain (z > 0), j=0: lower domain
    """
    Xn = x - Vxn * t + Sxn          # dependent Cartesian coordinate (Eq. 45)
    Zn = (-1)**j * Rzn * z           # decay variable (Eq. 46)
    
    cx_n = np.cos(Rzn * Xn)          # SKF cosine function (Eq. 41)
    sx_n = np.sin(Rzn * Xn)          # SKF sine function (Eq. 42)  
    ez_n = np.exp(Zn)                # SKE function (Eq. 44)
    
    return (Fhn * cx_n + Ghn * sx_n) * ez_n

# Evaluate at selected parameter values: z=1, t=1, Vxn=1, Sxn=1, Fhn=1, Ghn=1
params_paper = dict(z=1, t=1, Vxn=1, Sxn=1, Fhn=1, Ghn=1, j=1)

x_test = np.linspace(-1, 1, 5)
print("SKEF structure ah_n at paper parameters (z=1, t=1, Vxn=Sxn=Fhn=Ghn=1):")
print(f"{'x':<10} {'Rzn=4':<15} {'Rzn=7':<15} {'Rzn=11'}")
print("-" * 55)
for x in x_test:
    v4  = ah_n(x, **{k:v for k,v in params_paper.items()}, Rzn=4)
    v7  = ah_n(x, **{k:v for k,v in params_paper.items()}, Rzn=7)
    v11 = ah_n(x, **{k:v for k,v in params_paper.items()}, Rzn=11)
    print(f"  {x:<8.1f} {v4:<15.6f} {v7:<15.6f} {v11:.6f}")
```

## 5. Trigonometric Reduction of the Fundamental SKEF Structure $ah_n$

To explore the effect of fundamental parameters, we define trigonometric functions through structural coefficients $Fh_n$ and $Gh_n$:

$$\sin(A_n) = \frac{Fh_n}{\sqrt{Fh_n^2 + Gh_n^2}} \tag{12.1}, \quad \cos(A_n) = \frac{Gh_n}{\sqrt{Fh_n^2 + Gh_n^2}} \tag{12.2}$$

$$A_n = \arcsin\!\left(\frac{Fh_n}{\sqrt{Fh_n^2 + Gh_n^2}}\right) \tag{12.3}$$

Using the sine addition identity $\sin(A+B) = \sin(A)\cos(B) + \cos(A)\sin(B)$, the SKEF structure reduces to:

$$ah_n = \sqrt{Fh_n^2 + Gh_n^2}\;\sin(Rz_n\,X_n + A_n)\;ez_n \tag{12.7}$$

Substituting $ez_n = e^{(-1)^j Rz_n z}$, the **trigonometric form** becomes:

$$ah_n = A_n \sin\!\left(Rz_n(x - s_n)\right) \tag{13.2}$$

where in the upper domain ($j=1$, $z > 0$):

$$A_n = \sqrt{Fh_n^2 + Gh_n^2}\; e^{-Rz_n\,z} \tag{13.3}$$

$$s_n = Vx_n\,t - Sx_n - \frac{\arcsin\!\left(\dfrac{Fh_n}{\sqrt{Fh_n^2+Gh_n^2}}\right)}{Rz_n} \tag{13.4}$$


```python
# Verify the trigonometric reduction
# With paper parameters: z=5/6, t=3, Vzn=2/3, Sxn=-7/3, Fhn=5, Ghn=2 (from Eq. 13.5)

z_p   = 5/6
t_p   = 3
Vxn_p = 2/3
Sxn_p = -7/3
Fhn_p = 5
Ghn_p = 2

# Amplitude formula (Eq. 13.6 / 13.7):
# A_n = sqrt(Fhn^2 + Ghn^2) * exp(-Rzn * z)
# A_n = sqrt(29) * exp(-5*Rzn/6) = 5.385 * exp(-0.8333 * Rzn)

Rzn_range = np.linspace(0.01, 5, 400)
A_n = np.sqrt(Fhn_p**2 + Ghn_p**2) * np.exp(-Rzn_range * z_p)

print("Amplitude A_n — exponential form (Eq. 13.6–13.7):")
print(f"  sqrt(Fhn^2 + Ghn^2) = sqrt({Fhn_p**2} + {Ghn_p**2}) = sqrt(29) = {np.sqrt(29):.4f}")
print(f"  Decay coefficient: -z = -{z_p} = -5/6 ≈ -0.8333")
print(f"  A_n = {np.sqrt(29):.3f} * exp(-0.8333 * Rzn)  (Eq. 13.7)")
print()

# Shift formula (Eq. 13.4 / 16.1-16.2):
# s_n = Vxn*t - Sxn - arcsin(Fhn/sqrt(Fhn^2+Ghn^2)) / Rzn
# s_n = 13/3 - arcsin(5*sqrt(29)/29) / Rzn = 4.3333 - 1.1902/Rzn
An_phase = np.arcsin(Fhn_p / np.sqrt(Fhn_p**2 + Ghn_p**2))
s_n_const = Vxn_p * t_p - Sxn_p
print(f"Shift s_n verification (Eq. 16.1-16.2 from paper):")
print(f"  Vxn*t - Sxn = {Vxn_p}*{t_p} - ({Sxn_p}) = {s_n_const:.4f} = 13/3 ≈ 4.3333  ✓")
print(f"  arcsin(Fhn/sqrt(Fhn^2+Ghn^2)) = arcsin(5/sqrt(29)) = {An_phase:.4f}")
print(f"  s_n = 4.3333 - 1.1902/Rzn  (Eq. 16.2)")
```

## 6. Research on the Wave Parameter $Rz_n$ of the Fundamental SKEF Structure $ah_n$

We explore the effect of $Rz_n$ on:
- **Amplitude** $A_n$ — exponential function of $Rz_n$

- **Horizontal shift** $s_n$ — hyperbola in $Rz_n$

Fixed parameters (Eq. 13.5 from paper):
$$z = \frac{5}{6},\; t = 3,\; Vx_n = \frac{2}{3},\; Sx_n = -\frac{7}{3},\; Fh_n = 5,\; Gh_n = 2$$


```python
# ============================================================
# Plot 1: Amplitude A_n — exponential decay in Rz_n (Eq. 13.6–13.7)
# A_n = sqrt(29) * exp(-5*Rzn/6)
# ============================================================

Rzn_vals = np.linspace(0, 5, 400)
A_n_vals = np.sqrt(29) * np.exp(-5*Rzn_vals/6)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ax1 = axes[0]
ax1.plot(Rzn_vals, A_n_vals, color='#1F3864', linewidth=2.5)
ax1.fill_between(Rzn_vals, A_n_vals, alpha=0.1, color='#1F3864')
ax1.set_xlabel('$Rz_n$', fontsize=13)
ax1.set_ylabel('$A_n$', fontsize=13)
ax1.set_title('Amplitude $A_n$ vs $Rz_n$\n$A_n = \\sqrt{29}\\, e^{-5Rz_n/6} = 5.385\\, e^{-0.8333\\, Rz_n}$', fontsize=11)
ax1.set_xlim([0, 5])
ax1.set_ylim([0, 5.5])
ax1.annotate(f'y-intercept: $\sqrt{{29}} \\approx {np.sqrt(29):.3f}$',
             xy=(0, np.sqrt(29)), xytext=(1.5, 4.5),
             arrowprops=dict(arrowstyle='->', color='gray'),
             fontsize=10, color='#1F3864')
ax1.axhline(y=0, color='black', linewidth=0.8, alpha=0.5, linestyle='--',
            label='Horizontal asymptote $A_n=0$')
ax1.legend(fontsize=9)

# Confirm it's exponential: log(A_n) vs Rzn should be linear
ax2 = axes[1]
ax2.plot(Rzn_vals[1:], np.log(A_n_vals[1:]), color='#1F3864', linewidth=2.5)
ax2.set_xlabel('$Rz_n$', fontsize=13)
ax2.set_ylabel('$\ln(A_n)$', fontsize=13)
ax2.set_title('log-linear plot confirms exponential decay\n$\ln(A_n) = \\ln(\\sqrt{29}) - \\frac{5}{6}Rz_n$', fontsize=11)
ax2.set_xlim([0, 5])

plt.suptitle('Analysis 1: Exponential Decay of Amplitude $A_n$', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('amplitude_An.png', dpi=150, bbox_inches='tight')
plt.show()
print("Figure saved: amplitude_An.png")
print(f"At Rzn=0: A_n = sqrt(29) = {np.sqrt(29):.4f} (y-intercept, Eq. 14.1)")
print(f"As Rzn→∞: A_n → 0 (horizontal asymptote, Eq. 14.2)")
```

```python
# ============================================================
# Plot 2: Period P_n as a hyperbola in Rz_n (Eq. 15.1)
# P_n = 2*pi / Rzn
# ============================================================

Rzn_pos = np.linspace(0.05, 5, 400)
P_n = 2 * np.pi / Rzn_pos

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ax1 = axes[0]
ax1.plot(Rzn_pos, P_n, color='#2E7D32', linewidth=2.5)
ax1.fill_between(Rzn_pos, P_n, alpha=0.1, color='#2E7D32')
ax1.set_xlabel('$Rz_n$', fontsize=13)
ax1.set_ylabel('$P_n$', fontsize=13)
ax1.set_title('Period $P_n$ vs $Rz_n$\n$P_n = 2\\pi / Rz_n$ — Hyperbola', fontsize=11)
ax1.set_xlim([0, 5])
ax1.set_ylim([0, 45])
ax1.axhline(y=0, color='gray', linestyle='--', alpha=0.7, linewidth=1,
            label='$P_n \\to 0$ as $Rz_n \\to \\infty$ (Eq. 15.3)')
ax1.axvline(x=0, color='gray', linestyle=':', alpha=0.7, linewidth=1,
            label='$P_n \\to \\infty$ as $Rz_n \\to 0^+$ (Eq. 15.2)')
ax1.legend(fontsize=9)

# Show specific Rzn values from paper
ax2 = axes[1]
x = np.linspace(-1.5, 1.5, 500)
z_p, t_p, Vxn_p, Sxn_p = 1.0, 1.0, 1.0, 1.0
Fhn_p, Ghn_p = 1.0, 1.0

for Rzn, color, ls in [(4, '#1F3864', '-'), (7, '#C62828', '--'), (11, '#E65100', ':')]:
    vals = np.array([ah_n(xi, z_p, t_p, Rzn, Vxn_p, Sxn_p, Fhn_p, Ghn_p) for xi in x])
    Pn = 2*np.pi/Rzn
    ax2.plot(x, vals, color=color, linestyle=ls, linewidth=2,
             label=f'$Rz_n={Rzn}$, $P_n={Pn:.3f}$')

ax2.set_xlabel('$x$', fontsize=13)
ax2.set_ylabel('$ah_n$', fontsize=13)
ax2.set_title('SKEF structure at $Rz_n$ = 4, 7, 11\n(z=t=Vx=Sx=Fh=Gh=1, upper domain)', fontsize=10)
ax2.legend(fontsize=9)
ax2.axhline(y=0, color='black', linewidth=0.5)

plt.suptitle('Analysis 2: Hyperbolic Period $P_n = 2\\pi/Rz_n$', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('period_Pn.png', dpi=150, bbox_inches='tight')
plt.show()
print("Figure saved: period_Pn.png")
print(f"lim(Rzn→0+) P_n = ∞  (vertical asymptote)")
print(f"lim(Rzn→∞) P_n = 0   (horizontal asymptote)")
```

```python
# ============================================================
# Plot 3: Horizontal shift s_n as a hyperbola in Rz_n (Eq. 16.1–16.2)
# s_n = 4.3333 - 1.1902/Rzn
# ============================================================

Rzn_pos = np.linspace(0.05, 5, 400)

# Selected parameters: Vxn=2/3, t=3, Sxn=-7/3, Fhn=5, Ghn=2
Vxn_p2, t_p2, Sxn_p2, Fhn_p2, Ghn_p2 = 2/3, 3, -7/3, 5, 2
An_phase_p2 = np.arcsin(Fhn_p2 / np.sqrt(Fhn_p2**2 + Ghn_p2**2))  # ≈ 1.1902
s_const_p2  = Vxn_p2 * t_p2 - Sxn_p2  # = 13/3 ≈ 4.3333

s_n_vals = s_const_p2 - An_phase_p2 / Rzn_pos

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

ax1 = axes[0]
ax1.plot(Rzn_pos, s_n_vals, color='#C62828', linewidth=2.5)
ax1.set_xlabel('$Rz_n$', fontsize=13)
ax1.set_ylabel('$s_n$', fontsize=13)
ax1.set_title('Horizontal Shift $s_n$ vs $Rz_n$\n$s_n = 4.3333 - 1.1902/Rz_n$', fontsize=11)
ax1.set_xlim([0, 5])
ax1.set_ylim([-4.5, 4.5])
ax1.axhline(y=s_const_p2, color='gray', linestyle='--', alpha=0.7,
            label=f'$s_n \\to Vx_n t - Sx_n = {s_const_p2:.4f}$ as $Rz_n\\to\\infty$ (Eq. 16.3)')
ax1.axvline(x=0, color='gray', linestyle=':', alpha=0.5, linewidth=1)
ax1.legend(fontsize=8)
ax1.annotate(f'Horizontal asymptote\n$s_n = 13/3 \\approx {s_const_p2:.4f}$',
             xy=(4, s_const_p2), xytext=(2.5, 3.5),
             arrowprops=dict(arrowstyle='->', color='gray'), fontsize=9)

# Demonstration: effect of Rzn on x-shift of ah_n waveform
ax2 = axes[1]
x = np.linspace(0, 8, 500)
for Rzn, color in [(0.5, '#C62828'), (1.0, '#E65100'), (2.0, '#1565C0'), (2.5, '#2E7D32')]:
    sn = s_const_p2 - An_phase_p2/Rzn
    vals = np.array([ah_n(xi, z=5/6, t=3, Rzn=Rzn, Vxn=2/3, Sxn=-7/3,
                         Fhn=5, Ghn=2) for xi in x])
    # normalize for visibility
    mx = np.max(np.abs(vals))
    if mx > 0:
        vals = vals / mx
    ax2.plot(x, vals, color=color, linewidth=1.8,
             label=f'$Rz_n={Rzn}$, $s_n={sn:.2f}$')

ax2.set_xlabel('$x$', fontsize=13)
ax2.set_ylabel('$ah_n$ (normalized)', fontsize=13)
ax2.set_title('Phase shift demonstration\n(paper params: z=5/6, t=3, Fh=5, Gh=2)', fontsize=10)
ax2.legend(fontsize=9)
ax2.axhline(y=0, color='black', linewidth=0.5)

plt.suptitle('Analysis 3: Hyperbolic Horizontal Shift $s_n$', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('shift_sn.png', dpi=150, bbox_inches='tight')
plt.show()
print("Figure saved: shift_sn.png")
```

## Summary

The following results are obtained from computational analysis of the SKEF structural functions:

### Key Results

**1. Amplitude $A_n$ — Exponential decay (Eq. 13.6–13.7):**
$$A_n = \sqrt{Fh_n^2 + Gh_n^2}\; e^{-Rz_n z} = \sqrt{29}\; e^{-0.8333\, Rz_n}$$
Starting at y-intercept $\sqrt{Fh_n^2 + Gh_n^2}$ with horizontal asymptote $A_n = 0$ as $Rz_n \to \infty$.

**2. Period $P_n$ — Hyperbola (Eq. 15.1):**
$$P_n = \frac{2\pi}{Rz_n}$$
Vertical asymptote $Rz_n = 0$ and horizontal asymptote $P_n = 0$.

**3. Horizontal shift $s_n$ — Hyperbola (Eq. 16.1–16.2):**
$$s_n = 4.3333 - \frac{1.1902}{Rz_n}$$
Horizontal asymptote $s_n \to Vx_n t - Sx_n$ as $Rz_n \to \infty$.

**4. Combined effect (Eq. 17.1):** When $Rz_n$ increases from $1/2$ to $5/2$:
- $A_n$ decreases nearly **5 times**
- $P_n$ diminishes approximately **5 times**
- $s_n$ remains positive

These results are consistent with the theoretical analysis in Sections 1–5 and confirm that $Rz_n$ is the primary structural parameter controlling scalar field dynamics within the SKEF framework.
